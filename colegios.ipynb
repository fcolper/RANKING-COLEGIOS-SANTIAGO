```python
import pandas as pd
```

```python
csv_2023 = pd.read_csv('2025-2023.csv', sep=';')
csv_2020 = pd.read_csv('2022-2020.csv', sep=';')
csv_2018 = pd.read_csv('2019-2018.csv', sep=';')
```

```python
csv_2023_rm=csv_2023[csv_2023['CODIGO_REGION']==13.0]
df_filtrado_rm=csv_2023_rm[csv_2023_rm['SITUACION_EGRESO']==1]
df_filtrado_rm['Promedio_len_mat']=(df_filtrado_rm['MATE_ACTUAL']+df_filtrado_rm['CLEC_ACTUAL'])/2
```

```python
df_filtrado_año=df_filtrado_rm[df_filtrado_rm['AÑO']==2023]

df_rm_2023=df_filtrado_año.groupby('RBD', as_index=False).agg({
    'ESTABLECIMIENTO': 'first',
    'AÑO':'mean',
    'SITUACION_EGRESO':'sum',
    'CLEC_ACTUAL':'mean',
    'MATE_ACTUAL':'mean',
    'Promedio_len_mat':'mean',

    })

df_rm_2023['Ranking'] = df_rm_2023['Promedio_len_mat'].rank(method='dense', ascending=False).astype(int)
df_rm_2023['AÑO'].astype('Int64')


df_rm_2023.sort_values('Ranking')
```

```python
df_filtrado_año=df_filtrado_rm[df_filtrado_rm['AÑO']==2024]

df_rm_2024=df_filtrado_año.groupby('RBD', as_index=False).agg({
    'ESTABLECIMIENTO': 'first',
    'AÑO':'mean',
    'SITUACION_EGRESO':'sum',
    'CLEC_ACTUAL':'mean',
    'MATE_ACTUAL':'mean',
    'Promedio_len_mat':'mean',

    })

df_rm_2024['Ranking'] = df_rm_2024['Promedio_len_mat'].rank(method='dense', ascending=False).astype(int)
df_rm_2024['AÑO'].astype('Int64')


df_rm_2024.sort_values('Ranking')
```

```python
df_filtrado_año=df_filtrado_rm[df_filtrado_rm['AÑO']==2025]

df_rm_2025=df_filtrado_año.groupby('RBD', as_index=False).agg({
    'ESTABLECIMIENTO': 'first',
    'AÑO':'mean',
    'SITUACION_EGRESO':'sum',
    'CLEC_ACTUAL':'mean',
    'MATE_ACTUAL':'mean',
    'Promedio_len_mat':'mean',

    })

df_rm_2025['Ranking'] = df_rm_2025['Promedio_len_mat'].rank(method='dense', ascending=False).astype(int)
df_rm_2025['AÑO'].astype('Int64')


df_rm_2025.sort_values('Ranking')
```

```python
df_concat=pd.concat([df_rm_2023,df_rm_2024,df_rm_2025])
df_concat[df_concat['Ranking']==1]
```

```python
df_concat
```

```python
lat_lon_colegios=pd.read_csv('rbd_colegios.csv', sep=';')

df_colegios_2018_2025 = df_concat.merge(
    lat_lon_colegios[['RBD','NOM_COM_RBD','LATITUD', 'LONGITUD']], 
    how='left', 
    on='RBD'
)

#CAMBIAR COMA POR PUNTOS DE LA COLUMNA LATITUD y transformarlo a un float
df_colegios_2018_2025['LATITUD'] = df_colegios_2018_2025['LATITUD'].str.replace(',', '.')
df_colegios_2018_2025['LONGITUD'] = df_colegios_2018_2025['LONGITUD'].str.replace(',', '.')
# Intentar convertir a número
df_colegios_2018_2025["LATITUD"] = pd.to_numeric(df_colegios_2018_2025["LATITUD"], errors="coerce")
df_colegios_2018_2025["LONGITUD"] = pd.to_numeric(df_colegios_2018_2025["LONGITUD"], errors="coerce")

df_colegios_2018_2025
```

```python
import plotly.express as px

# Leer el DataFrame (suponiendo que ya está cargado como df)
# df = pd.read_csv("archivo.csv")  # Si se necesita cargar desde un archivo

colegios = ["COLEGIO SAINT GASPAR COLLEGE", "COLEGIO THE ENGLISH INSTITUTE", "COLEGIO SAN IGNACIO EL BOSQUE",
            'COLEGIO SAINT GABRIEL´S SCHOOL LOCAL SEDE', 'COLEGIO TREWHELA´S ENGLISH SCHOOL', 'COLEGIO AKROS',
            'COLEGIO SUIZO DE SANTIAGO','COLEGIO PUMAHUE', 'ESCUELA BASICA N°2145 COLEGIO DUNALASTAIR PEÑALOLÉN',
            'COLEGIO AMERICAN BRITISH COLLEGE', 'COLEGIO SANTA MARIA DE LO CANAS','COLEGIO ANDREE ENGLISH SCHOOL',
            'COLEGIO DE LA SALLE']
df_colegio=df_colegios_2018_2025[df_colegios_2018_2025['ESTABLECIMIENTO'].isin(colegios)]

# Ordenar el DataFrame por AÑO
df = df_colegio.sort_values(by="AÑO")

# Crear el gráfico
fig = px.line(df, x="AÑO", y="Ranking", color="ESTABLECIMIENTO", markers=True, 
              title="Evolución del Ranking de Establecimientos",
              labels={"AÑO": "Año", "Ranking": "Ranking"},
              color_discrete_sequence=px.colors.qualitative.Set1,
              hover_data={"Promedio_len_mat": True})

#REVERSE EJE Y
fig.update_yaxes(autorange="reversed")

# Mostrar el gráfico
fig.show()
```

```python
import plotly.express as px

# Leer el DataFrame (suponiendo que ya está cargado como df)
# df = pd.read_csv("archivo.csv")  # Si se necesita cargar desde un archivo

comunas = ['PEÑALOLÉN','LA REINA', 'LA FLORIDA', 'ÑUÑOA', 'PROVIDENCIA']

df_colegio_comunas=df_colegios_2018_2025[df_colegios_2018_2025['NOM_COM_RBD'].isin(comunas)]
df_colegio_comunas_100=df_colegio_comunas[df_colegio_comunas['Ranking']<150]

# Ordenar el DataFrame por AÑO
df = df_colegio_comunas_100.sort_values(by="AÑO")

# Crear el gráfico
fig = px.line(df, x="AÑO", y="Ranking", color="ESTABLECIMIENTO", markers=True, 
              title="Evolución del Ranking de Establecimientos",
              labels={"AÑO": "Año", "Ranking": "Ranking"},
              color_discrete_sequence=px.colors.qualitative.Set1,
              hover_data={"Promedio_len_mat": ":.2f"})

#REVERSE EJE Y
fig.update_yaxes(autorange="reversed")

# Mostrar el gráfico
fig.show()
```

```python
# Eliminar filas con comuna desconocida
df = df_colegios_2018_2025[df_colegios_2018_2025["NOM_COM_RBD"] != "#N/D"]

# Agrupar por comuna y año
df_grouped = df.groupby(["AÑO", "NOM_COM_RBD"])["Promedio_len_mat"].mean().reset_index()

# Graficar la evolución del rendimiento por comuna con Plotly Express
fig = px.line(
    df_grouped,
    x="AÑO",
    y="Promedio_len_mat",
    color="NOM_COM_RBD",
    markers=True,
    title="Evolución del puntaje promedio por comuna",
    labels={"AÑO": "Año", "Promedio_len_mat": "Promedio Lenguaje y Matemática", "NOM_COM_RBD": "Comuna"},
)

fig.update_layout(legend_title_text="Comuna", template="plotly_dark")
fig.show()
```

```python
df_colegios_2025=df_colegios_2018_2025[df_colegios_2018_2025['AÑO']==2025]

df_colegios_2025["size_inverse"] = df_colegios_2025["Ranking"].max() - df_colegios_2025["Ranking"] + 1

fig = px.scatter_mapbox(
    df_colegios_2025,
    lat="LATITUD",
    lon="LONGITUD",
    center={"lat": -33.45, "lon": -70.65},
    color="Promedio_len_mat",  # Color según el promedio
    size="size_inverse",       # Tamaño inversamente proporcional al ranking
    size_max=10,               # Tamaño máximo del círculo
    zoom=10,
    mapbox_style="carto-positron",
    title="Mapa de Colegios 2025",
    height=600,
    hover_data={
        "Ranking": True, 
        "ESTABLECIMIENTO": True, 
        "Promedio_len_mat": ":.1f",
        "LATITUD": False,  # Ocultamos estos datos
        "LONGITUD": False,
        'size_inverse': False
    }
)
fig.show()
```
